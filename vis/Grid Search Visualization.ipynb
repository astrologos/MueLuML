```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# Read in merged grid search files
df1 = pd.read_csv('./data/grid_search_merged.csv', header=None)
df1.columns = [int(d*10) + 1 for d in np.arange(0, 1, 0.01)]
df1.index = [d for d in np.arange(0, 0.99, 0.01)]
df1.head()
```

```python
ax = sns.heatmap(df1)
plt.xticks(rotation=0)
plt.xlabel('Chebshev Degree')
plt.ylabel('Drop Tolerance')
plt.savefig('grid_search.png', dpi=100)
plt.show()
```

```python
# Read in merged grid search files
df2 = pd.read_csv('./data/droptol_coarsesize/grid_search_merged.csv', header=None)
df2.columns = [int(d*1980) + 20 for d in np.arange(0, 1.04, 0.04)]
df2.index = [d for d in np.arange(0, 0.96, 0.04)]
df2.head()
```
